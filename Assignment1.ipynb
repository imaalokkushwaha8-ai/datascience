```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

# LINE CHART

```python
df=pd.read_csv("Health_heart_experimental.csv")
```

```python
age = [64,21,30,35,39,68,50,23,34,36]
HR = [128,106,126,130,112,130,89,131,90,187]
# sorting age
sorted_pairs = sorted(zip(age, HR))
age_sorted, HR_sorted = zip(*sorted_pairs)

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(age_sorted,HR_sorted, color='green', marker='+', linestyle='dotted')

# add a title
plt.title("HR of different age groups")

# add a label to the y-axisYears
plt.ylabel("HR")
plt.xlabel("age")
plt.show()
```

## FINDINGS FROM LINE CHART(HR VS AGE)
The line chart shows how heart rate changes as age increases. The values go up and down at different ages, meaning heart rate does not steadily increase or decrease with age.

```python
# create dataframe from Health-heart_experimental.CSV
df = pd.DataFrame({
    "weightKg": [69,48,69,45,92,53,56,55,67,46],
    "heart_rate": [128,106,126,130,112,130,89,131,90,187]
})

# group by weightKg and take mean
df_new = df.groupby('weightKg').mean(numeric_only=True).reset_index()

# line chart
sns.lineplot(x='weightKg', y='heart_rate', data=df_new)

plt.title("Average Heart Rate by WeightKg")
plt.show()
```

## FINDINGS FROM LINE CHART2(HR VS WEIGHTKG)
The code groups the data by weight (kg), calculates the average heart rate for each weight, and then draws a line chart showing how heart rate changes with different body weights.

# BAR PLOT

```python
weightKg=[69,48,69,45,92,53,56,55,67,46]
heightCm=[147,150,183,149,166,172,172,169,178,114]

#plot bars with weightKg in x-axis and heightCm in y-axis
plt.bar(weightKg,heightCm)

#Add a title
plt.title("Average HeightCm by WeightKg")

#label the y-axis
plt.ylabel("heightCm")
#label the y-axis
plt.xlabel("weightKg")

plt.show()
```

## FINDINGS FROM BAR PLOT
The bar plot shows how weight varies with height, with taller individuals generally having higher weights, though there is some variation.
