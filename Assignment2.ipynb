## MATRIX OPERATION

```python
import numpy as np
```

```python
A=np.array([[2,-1],[0,3]])
B=np.array([[1,2],[-3,4]])
print("Matrix A is ")
print(A)
print("Matrix B is")
print(B)
print("The sum of A and B")
print(A+B)
print("The product of A and B")
print(A @ B)
print("Result of 3A-2B")
print(3*A-2*B)
```

## Determinant of a Matrix

```python
import numpy.linalg as LA
```

```python
A=np.array([[2,-1],[0,3]])
LA.det(A)
```

## Inverse of a matrix

```python
if(LA.det(A)!=0):
    print("A is invertible")
    A_inv=np.linalg.inv(A)
    print("Inverse of matrix A ")
    print(A_inv)
else:
    print("A is not invertible")
```

## Eigen Values And Eigen Vectors

```python
C=np.array([[4,1],[-2,1]])
print("Matrix C is")
print(C)
eigenvalues,eigenvectors=np.linalg.eig(A)
print("Eigenvalue(first)")
print(eigenvalues[0])
print("Corresponding eigenvector is ")
print(eigenvectors[:, 0])
print("Eigenvalue(second)")
print(eigenvalues[1])
print("Corresponding eigenvector is ")
print(eigenvectors[:, 1])
```

# To calculate the price of a car 

```python
print("The Feature Matrix is")
X=np.array([
    [4,1.5,100],
    [6,2,150]
])
print(X)
print("Price Vector")
y=np.array([
    [20],
    [30]
])
print(y)
# computing coefficient using pseudo inverse
beta=np.linalg.pinv(X).dot(y)
#To find the price of car3
print("Feature Matrix of car3")
X1=np.array([
    [8,3,200]
])
print(X1)
priceofcar3= X1 @ beta
print("The price of car 3 is")
print(priceofcar3)
```
